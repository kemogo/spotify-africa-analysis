# Spotify Africa Dataset - Feature Distribution Analysis

This notebook explores and visualizes the distribution of features in the Spotify Africa Dataset from Hugging Face.

## 1. Import Required Libraries

```python
# Import pandas - a library for working with data in tables (like Excel)
import pandas as pd

# Import numpy - a library for mathematical operations on numbers
import numpy as np

# Import matplotlib - a library for creating static charts and graphs
import matplotlib.pyplot as plt

# Import seaborn - makes matplotlib graphs look prettier and easier to create
import seaborn as sns

# Import plotly - a library for creating interactive charts you can hover over
import plotly.express as px
import plotly.graph_objects as go

# Import load_dataset - a function to download datasets from Hugging Face
from datasets import load_dataset

# Import warnings - helps us hide warning messages that aren't important
import warnings
warnings.filterwarnings('ignore')  # This hides all warning messages

# Set the visual style for our graphs to 'whitegrid' (white background with grid lines)
sns.set_style('whitegrid')

# Set the default size for all our graphs (12 inches wide, 6 inches tall)
plt.rcParams['figure.figsize'] = (12, 6)
```

## 2. Load the Spotify Africa Dataset

```python
# Step 1: Print a message to let us know the loading has started
print("Loading Spotify Africa Dataset (master_tracks)...")

# Step 2: Import the load_dataset function (in case it wasn't imported earlier)
from datasets import load_dataset

# Step 3: Download and load the dataset from Hugging Face
# - "electricsheepafrica/Spotify-Africa-Dataset" is the dataset name on Hugging Face
# - data_files specifies we only want the master_tracks folder (not all 20 datasets)
# - *.parquet means "all files ending with .parquet" in that folder
dataset = load_dataset(
    "electricsheepafrica/Spotify-Africa-Dataset",  # Dataset location
    data_files="data/datasets/master_tracks/*.parquet"  # Specific file to load
)

# Step 4: Convert the dataset to a pandas DataFrame
# - DataFrame is like an Excel spreadsheet - rows and columns of data
# - dataset['train'] gets the training data portion
# - pd.DataFrame() converts it to a format we can easily work with
df = pd.DataFrame(dataset['train'])

# Step 5: Print success message and show information about the data
print(f"✓ Dataset loaded successfully!")  # The ✓ is a checkmark symbol

# Show how many rows and columns we have
# - df.shape[0] = number of rows (tracks in our dataset)
# - df.shape[1] = number of columns (features/information about each track)
print(f"Shape: {df.shape[0]} rows × {df.shape[1]} columns")

# Step 6: Display the column names (what information we have about each track)
print(f"\nColumn names:")
print(list(df.columns))  # list() converts column names to a readable list

# Step 7: Show the first 5 rows of data so we can see what it looks like
print(f"\nFirst few rows:")
df.head()  # head() shows the first 5 rows by default
```

## 3. Dataset Overview

```python
# Display basic information about our dataset
print("Dataset Information:")
print("=" * 50)  # Print a line of 50 equal signs for visual separation

# df.info() shows:
# - Column names
# - How many non-null (non-empty) values in each column
# - Data type of each column (number, text, etc.)
df.info()

print("\n" + "=" * 50)  # \n creates a new line
print("\nBasic Statistics:")
print("=" * 50)

# df.describe() calculates statistics for numerical columns:
# - count: how many values
# - mean: average value
# - std: standard deviation (how spread out the values are)
# - min: smallest value
# - 25%, 50%, 75%: quartiles (percentiles)
# - max: largest value
df.describe()
```

## 4. Check for Missing Values

```python
# Check for missing (null/empty) values in our dataset

# Step 1: Count how many missing values are in each column
# - df.isnull() returns True for empty cells, False for filled cells
# - .sum() counts how many True values (missing values) in each column
missing_values = df.isnull().sum()

# Step 2: Calculate what percentage of each column is missing
# - len(df) gives us the total number of rows
# - We divide missing values by total rows and multiply by 100 to get percentage
missing_percent = (missing_values / len(df)) * 100

# Step 3: Create a new DataFrame to organize our results
# - This creates a table with two columns: 'Missing Values' and 'Percentage'
missing_df = pd.DataFrame({
    'Missing Values': missing_values,  # Number of missing values
    'Percentage': missing_percent       # Percentage missing
})

# Step 4: Filter to show only columns that have missing values
# - missing_df['Missing Values'] > 0 finds columns with at least 1 missing value
# - .sort_values() sorts from most missing to least missing
# - ascending=False means largest values first
missing_df = missing_df[missing_df['Missing Values'] > 0].sort_values('Missing Values', ascending=False)

# Step 5: Display results
# - if len(missing_df) > 0 means "if there are any columns with missing values"
if len(missing_df) > 0:
    print("Missing Values Summary:")
    print(missing_df)
else:
    print("No missing values found in the dataset!")
```

## 5. Feature Distribution Analysis

Let's visualize the distribution of numerical features in the dataset.

```python
# Identify which columns contain numbers (vs. text)

# df.select_dtypes() filters columns by their data type
# - include=[np.number] means "only include columns with numbers"
# - This gives us things like: age, price, duration, popularity, etc.
# - It excludes text columns like: name, artist, genre, etc.
# .columns gets the column names
# .tolist() converts them to a regular Python list
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()

# Print what we found
print(f"Numerical Features: {numerical_cols}")  # Show the list of column names
print(f"Total: {len(numerical_cols)} features")  # len() counts how many items in the list
```

### 5.1 Distribution Plots for All Numerical Features

```python
# Create distribution plots (histograms) for all numerical features

# Step 1: Calculate how many rows and columns of graphs we need
n_cols = 3  # We want 3 graphs per row
# Calculate rows needed: divide total features by 3, round up
# (len(numerical_cols) + n_cols - 1) ensures we round up
n_rows = (len(numerical_cols) + n_cols - 1) // n_cols  # // is integer division

# Step 2: Create a figure with multiple subplots (small graphs arranged in a grid)
# - figsize=(18, 5 * n_rows) sets the size: 18 inches wide, height based on rows
fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, 5 * n_rows))

# Step 3: Flatten the axes array so we can loop through it easily
# - If we have multiple rows, axes is a 2D array
# - .flatten() makes it a simple 1D list
axes = axes.flatten() if n_rows > 1 else [axes]

# Step 4: Loop through each numerical column and create a histogram
for idx, col in enumerate(numerical_cols):  # enumerate gives us index and column name
    if idx < len(axes):  # Make sure we don't run out of subplot spaces
        # Create a histogram with a smooth curve (KDE = Kernel Density Estimate)
        # - df[col] gets the data for this column
        # - kde=True adds the smooth curve on top of bars
        # - ax=axes[idx] tells it which subplot to use
        # - color='steelblue' makes the bars blue
        sns.histplot(df[col], kde=True, ax=axes[idx], color='steelblue')
        
        # Set the title for this subplot
        axes[idx].set_title(f'Distribution of {col}', fontsize=12, fontweight='bold')
        
        # Label the x-axis (horizontal) with the column name
        axes[idx].set_xlabel(col)
        
        # Label the y-axis (vertical) as "Frequency" (how often values appear)
        axes[idx].set_ylabel('Frequency')

# Step 5: Hide any extra subplots we don't need
# - If we have 8 features but created space for 9, hide the 9th
for idx in range(len(numerical_cols), len(axes)):
    axes[idx].set_visible(False)

# Step 6: Adjust spacing between subplots so they don't overlap
plt.tight_layout()

# Step 7: Display the final figure
plt.show()
```

**What This Shows:**

These histograms reveal the distribution patterns of numerical features:
- **Skewness**: Whether data clusters on the left (right-skewed) or right (left-skewed)
- **Peaks**: Most common values in the dataset (shown by the highest bars)
- **Spread**: How widely values are distributed across the range
- **Outliers**: Unusual values that stand apart from the main distribution

For example, if `popularity` is right-skewed, it means most tracks have low popularity with few highly popular tracks.

### 5.2 Box Plots for Outlier Detection

```python
# Create box plots for outlier detection
fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, 5 * n_rows))
axes = axes.flatten() if n_rows > 1 else [axes]

for idx, col in enumerate(numerical_cols):
    if idx < len(axes):
        sns.boxplot(data=df, y=col, ax=axes[idx], color='lightcoral')
        axes[idx].set_title(f'Box Plot of {col}', fontsize=12, fontweight='bold')
        axes[idx].set_ylabel(col)

# Hide extra subplots
for idx in range(len(numerical_cols), len(axes)):
    axes[idx].set_visible(False)

plt.tight_layout()
plt.show()
```

**What This Shows:**

Box plots help identify outliers and data spread:
- **Box**: Contains the middle 50% of data (from 25th to 75th percentile)
- **Line inside box**: Median value (50th percentile)
- **Whiskers**: Extend to show the typical range of data
- **Dots beyond whiskers**: Outliers - unusual values that may need investigation

For example, if `duration_ms` has many outlier dots, it indicates some tracks are unusually long or short compared to typical songs.

### 5.3 Correlation Heatmap

```python
# Create a correlation heatmap to see relationships between numerical features

# Step 1: Create a new figure with specific size
plt.figure(figsize=(14, 10))  # 14 inches wide, 10 inches tall

# Step 2: Calculate correlations between all numerical columns
# - Correlation measures how two variables move together
# - Values range from -1 to +1:
#   * +1 = perfect positive relationship (both increase together)
#   * 0 = no relationship
#   * -1 = perfect negative relationship (one increases, other decreases)
# df[numerical_cols] selects only the numerical columns
# .corr() calculates correlation between each pair of columns
correlation_matrix = df[numerical_cols].corr()

# Step 3: Create a heatmap to visualize the correlation matrix
# - annot=True displays the correlation numbers on each cell
# - fmt='.2f' formats numbers to 2 decimal places (e.g., 0.85)
# - cmap='coolwarm' uses blue for negative, red for positive correlations
# - center=0 centers the color scale at 0
# - square=True makes each cell a square
# - linewidths=1 adds thin white lines between cells
# - cbar_kws={"shrink": 0.8} makes the color bar 80% of normal size
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            center=0, square=True, linewidths=1, cbar_kws={"shrink": 0.8})

# Step 4: Add a title to the plot
plt.title('Correlation Heatmap of Numerical Features', fontsize=16, fontweight='bold')

# Step 5: Adjust layout so nothing gets cut off
plt.tight_layout()

# Step 6: Display the heatmap
plt.show()
```

**What This Shows:**

The correlation heatmap reveals relationships between numerical features:
- **Red cells (positive correlation)**: When one feature increases, the other tends to increase
  - Example: `artist_followers` and `artist_popularity` likely correlate positively
- **Blue cells (negative correlation)**: When one feature increases, the other tends to decrease
  - Example: `track_age_years` and `is_recent` would correlate negatively
- **White/pale cells (near 0)**: No meaningful relationship between features
- **Diagonal (value 1.00)**: Each feature perfectly correlates with itself

This helps identify which features move together and which are independent, useful for understanding data patterns and selecting features for machine learning models.

### 5.4 Interactive Distribution Plots with Plotly

```python
# Create interactive histograms using Plotly
for col in numerical_cols:
    fig = px.histogram(df, x=col, nbins=50, 
                       title=f'Interactive Distribution of {col}',
                       labels={col: col},
                       color_discrete_sequence=['steelblue'])
    
    fig.update_layout(
        xaxis_title=col,
        yaxis_title='Frequency',
        showlegend=False,
        height=400
    )
    
    fig.show()
```

**What This Shows:**

Interactive histograms provide the same distribution insights as static plots, but with added benefits:
- **Hover tooltips**: See exact values and counts by hovering over bars
- **Zoom capability**: Focus on specific ranges of interest
- **Pan functionality**: Explore different sections of the distribution
- **Better for exploration**: Ideal for detailed investigation of patterns

These plots are especially useful when presenting findings to others or when you need to examine specific value ranges in detail.

## 6. Categorical Features Analysis

```python
# Identify which columns contain text/categories (vs. numbers)

# df.select_dtypes() filters columns by their data type
# - include=['object'] means "only include columns with text/categories"
# - This gives us things like: artist name, genre, country, etc.
# - It excludes numerical columns like: duration, popularity, etc.
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

# Print what we found
print(f"Categorical Features: {categorical_cols}")
print(f"Total: {len(categorical_cols)} features")

# Display value counts for each categorical feature
# This shows us the most common values in each text column
for col in categorical_cols:
    print(f"\n{col}:")  # Print the column name
    print("-" * 50)    # Print a separator line
    
    # .value_counts() counts how many times each unique value appears
    # .head(10) shows only the top 10 most common values
    print(df[col].value_counts().head(10))
```

### 6.1 Bar Charts for Categorical Features

```python
# Create bar charts for categorical features (top 20 values)
for col in categorical_cols:
    plt.figure(figsize=(14, 6))
    
    top_values = df[col].value_counts().head(20)
    
    sns.barplot(x=top_values.values, y=top_values.index, palette='viridis')
    plt.title(f'Top 20 Values for {col}', fontsize=14, fontweight='bold')
    plt.xlabel('Count')
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()
```

**What This Shows:**

Bar charts for categorical features reveal the frequency distribution of text/category values:
- **Longest bars**: Most common categories (e.g., most popular artists or genres)
- **Short bars**: Rare categories that appear infrequently in the dataset
- **Bar count**: Shows diversity - many bars indicate high variety, few bars indicate concentration
- **Dominance patterns**: Identifies if one or two values dominate the category

For example, if the `country` bar chart shows Nigeria with the longest bar, it means the dataset contains more Nigerian tracks than tracks from other countries. This helps understand dataset composition and potential biases.

## 7. Summary Statistics by Category

```python
# Calculate summary statistics grouped by meaningful categorical features

# Select a few interesting categorical columns to group by
# We'll avoid columns with too many unique values (like track_id) or too much missing data
interesting_categories = ['country', 'release_era', 'popularity_tier', 'market_scope', 'duration_category']

# For each categorical column, show how numerical features vary across categories
for cat_col in interesting_categories:
    print(f"\n{'='*80}")
    print(f"Average Numerical Features by {cat_col}:")
    print('='*80)
    
    # Group by the categorical column and calculate mean of numerical columns
    grouped_stats = df.groupby(cat_col)[numerical_cols].mean().round(2)
    
    # Display the grouped statistics
    print(grouped_stats)
    print()
```
